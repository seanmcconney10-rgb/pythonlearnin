```python
filename = "journal.txt"
while True:
  choice = input("(W)rite, (R)ead, or (Q)uit? ").strip().upper()

  if choice == "W":
    entry = input("Enter your entry: ")
    with open(filename, "a") as file:
      file.write(entry + "\n")
      print("entry saved.\n")

  elif choice == "R":
    print("--- your journal ---")
    try:
      with open(filename, "r") as file:
        for line in file:
          print(line.strip())
    except FileNotFoundError:
      print("no entries yet.")
    print()

  elif choice == "Q":
    print("Goodbye!")
    break

  else:
    print("invalid option. please choose W, R, or Q.\n")
```
