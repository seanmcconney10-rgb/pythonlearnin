```python
correct_pin = "1234"
balance = 500.0

# PIN verification loop
while True:
    pin = input("enter PIN: ")

    if pin == correct_pin:
        print("PIN verified.")
        break  # Exit the loop on successful PIN
    else:
        print("Incorrect PIN. Try again.\n")

# Withdrawal selection - This part executes ONLY after the PIN has been verified and the loop has broken.
amount = float(input("Enter Withdrawal amount: "))

if amount <= balance:
    balance -= amount
    print("Withdrawal successful.")
    print(f"Current balance: ${balance}")
else:
    print("Insufficient funds.")
```
