```python
# get user input
classification = input("enter classification code (B, S, L): ").upper()
start_odometer = int(input("start odometer: "))
end_odometer = int(input("end odometer: "))
days_rented = int(input("days rented:"))

# Calculate miles
total_miles = end_odometer - start_odometer

# Pricing rules (example)
if classification == "B" :
  daily_rate = 40
  free_miles = 100 * days_rented
  cost_per_mile = 0.25
  class_name = "Basic"
elif classification == "S" :
  daily_rate = 60
  free_miles = 100 * days_rented
  cost_per_mile = 0.35
  class_name = "luxury"
else:
  print("invalid classification")
  exit()

#Calculate charged miles
charged_miles = max(0, total_miles - free_miles)

#calculate bill
total_bill = (daily_rate * days_rented) + (charged_miles * cost_per_mile)

#Output summary
print("\n--- RENTAL SUMMARY ---")
print("Classification:", class_name)
print("Total Miles:", total_miles)
print("Charged Miles:", charged_miles)
print("Total Bill: $", total_bill)
```
