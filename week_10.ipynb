```python
players = []

# Input loop
while True:
  name = input("enter player name (or 'quit' to stop):")

  if name.lower() == "quit":
    break

  players.append(name)

#search
search_name = input("\nwgo would you like to search for? ")
if search_name in players:
  print("Result: Player found!")
else:
  print("Result: Player not found.")
```
